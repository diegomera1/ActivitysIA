# CON LOS DATOS DE CANCIONES, ELIJA LAS 3 COLUMNAS QUE USTED DESEE Y ORDENELAS ASCENDENTEMENTE

```python
import pandas as pd

data = pd.read_csv('musica.csv')
```

## Ordernar musica por artistas

```python
data.sort_values('artists', ascending = True)
```

## Ordernar musica por nombre

```python
data.sort_values('name', ascending = True)
```

## Ordernar musica por energia

```python
data.sort_values('energy', ascending = True)
```
